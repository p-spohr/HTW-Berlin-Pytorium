### Data prepocessing


#### 2. Bereinigung des Datensatzes

Ist der Datensatz richtig formatiert?

Gibt es Lücken im Datensatz?

- In Daten gibt es eine unnötige Spalte, heißt 'Unnamed: 0' und Zeile wie zum Bsp 95 Luxembourg hat keine Informationen von Jahr 2016-2019

```python
### Data prepocessing
df = df.drop(['Unnamed: 0', 'Rank'], axis = 1)
df_cleaned = df.dropna()

### reset index after drop rows
df_cleaned = df_cleaned.reset_index(drop = True)
df_cleaned
```

```python
### Transpose df_cleaned_neu
df_trans = df_cleaned.transpose()
df_trans
```

```python
### select first row in df_trans (Which contains countries)
country_spalte = df_trans.iloc[0]
country_spalte
```

```python
### rename columns 
df_trans.columns = country_spalte
df_trans
```

```python
### Drop doppelte 
df_trans = df_trans.drop('Country', axis = 0)
df_trans
```

```python
### Convert index format in df_trans
df_trans.index = pd.to_datetime(df_trans.index, format = '%Y')

### sort 
sort_df = df_trans.sort_index( axis = 0, ascending = True) 

### Select first 5 Columns
sort_df.iloc[:, :5]
```

#### 4. Übersichtliche Darstellung

Was kann man plotten, um das Argument klar zu machen?

### Total global production per year


```python
# Total global production per year
global_production = df_cleaned[['2016', '2017', '2018', '2019', '2020']].sum()


plt.plot(global_production.index, global_production.values, marker='o', label='Global Production')
plt.title('Global Soybean Production (2016-2020)')
plt.xlabel('Year')
plt.ylabel('Production (in tons)')
plt.grid(True)
plt.legend()
plt.show()
```

### Soybean production trends for each producer


```python
### Soybean production trends for each producer
def country_production(country):
    country_data = df_cleaned[df_cleaned['Country'] == country]

    plt.plot(['2016', '2017', '2018', '2019', '2020'], 
             country_data[['2016', '2017', '2018', '2019', '2020']].values.flatten(),
             marker='o', label=country)
    
    plt.title(f'Soybean Production Trends for {country} (2016-2020)')
    plt.xlabel('Year')
    plt.ylabel('Production (in tons)')
    plt.grid(True)
    plt.legend()
    plt.show()

widgets.interact(country_production, country=list(df_cleaned['Country'].values));
```

### Top 5 producer for each year


```python
### Here in 2017 top 5 but in 2018, paraguay is in top 5
### ->>> Problem, just show 5 familiar country ->>> Solve by put top_countries into function
```

```python
### Top 5 producer for each year
years = [ '2020', '2019', '2018', '2017','2016']


def top_production(year):
    top_countries = df_cleaned.nlargest(5, year)
    top_countries['Country']
    fig, ax = plt.subplots()
    x = top_countries['Country']
    y = top_countries[year]
    fig.suptitle(f'Top 5 Soybean Producing Countries in {year}')
    
    ax.bar(x, y)



radio_buttons = widgets.RadioButtons(
    options = years,
    description = 'Years' 
)

widgets.interact(top_production, year = radio_buttons);
```

### Producing differenz in percent of top 5 countries

```python
#### Diff methode in percent
sort_df_diff = sort_df.iloc[:, :5].pct_change()  ###### % change

#### drop 1. row in df, because 1. row contain NA 
sort_df_diff.drop(index= sort_df_diff.index[0], axis=0, inplace=True)
sort_df_diff
```

```python
###  plot % diff of soybean production in 5 largest producer

def percent_diff(producer):
    fig, ax = plt.subplots()
    
    # Plot production data for Brazil
    ax.plot(sort_df_diff.index.year, sort_df_diff[producer], marker='o', label= producer)
    
    #Add labels, title, and grid
    ax.set_title(f'Production in {producer} Over Time')
    ax.set_xlabel('Year')
    ax.set_ylabel('Production Change (%)')

    #Set x-axis ticks to display only specific years
    ax.set_xticks([2017, 2018, 2019, 2020])

    ax.grid(True)
    ax.legend()
    
    # Show the plot
    plt.show()

widgets.interact(percent_diff, producer = sort_df_diff.columns);
```
